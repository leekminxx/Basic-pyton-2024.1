## Folium 계속

```python
## file : test33_folium.ipynb
## desc : Folium 지도
import folium
```

```python
map = folium.Map(location=[35.173999, 129.196649], zoom_start=15, tiles='cartodb positron')
folium.Marker([35.173999, 129.196649], popup='Songjeong').add_to(map) # 마커 추가
map
```

```python
## 마커 종류 변경
map = folium.Map(location=[35.173999, 129.196649], zoom_start=15, tiles='cartodb positron')
folium.Marker([35.173999, 129.196649], popup='Songjeong', 
              icon=folium.Icon(color='red', icon='remove-circle')).add_to(map) # 마커 추가 + icon = 마커 변경
map
```
